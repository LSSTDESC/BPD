```python
from bpd.io import load_dataset
from arviz import ess
from blackjax.diagnostics import potential_scale_reduction

import numpy as np
```

```python
paths = [
    "/pscratch/sd/i/imendoza/data/cache_chains/exp73_51/shear_samples_5121_plus.npz",
    "/pscratch/sd/i/imendoza/data/cache_chains/exp73_51/shear_samples_5122_plus.npz",
    "/pscratch/sd/i/imendoza/data/cache_chains/exp73_51/shear_samples_5123_plus.npz",
    "/pscratch/sd/i/imendoza/data/cache_chains/exp73_51/shear_samples_5124_plus.npz",
]
```

```python
ds = {} 
for p in paths:
    ds1 = load_dataset(p)
    for k in ds1['samples']:
        if k not in ds:
            ds[k] = [ ds1['samples'][k] ]
        else:
            ds[k].append( ds1['samples'][k] )
```

```python
out = {} 
for k in ds:
    out[k] = np.stack(ds[k])
```

```python
# ess
out_ess = ess(out)
for k in out_ess:
    print(k)
    print(out_ess[k].data.item() / 4000)
    print()
    
```

```python
# rhat
out_ess = ess(out)
for k in out_ess:
    print(k)
    print(potential_scale_reduction(out[k]))
    print()
```
