```python
g1ps = ds['plus']['g1']
g1ps.shape
```

```python
plt.hist(g1ps[0, :], bins=21, histtype='step');
plt.hist(g1ps[1, :], bins=21, histtype='step');
```

```python
g1ps = ds['plus']['g1']
g2ps = ds['plus']['g2']
g1ms = ds['minus']['g1']
g2ms = ds['minus']['g2']


fig, axs = plt.subplots(2,2,figsize=(10,10))
get_pc_fig(axs[0,0], g1ps, 0.02, 'g1 (+)')
get_pc_fig(axs[0,1], g1ms, -0.02, 'g1 (-)')
get_pc_fig(axs[1,0], g2ps, 0.0, 'g2 (+)')
get_pc_fig(axs[1,1], g2ms, 0.0, 'g2 (-)')
plt.tight_layout()
```
