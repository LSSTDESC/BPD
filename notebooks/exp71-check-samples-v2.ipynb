# Alternative method of estimating errors

```python
from bpd.io import load_dataset
import numpy as np
import matplotlib.pyplot as plt 
```

```python
fpath = "/pscratch/sd/i/imendoza/data/cache_chains/exp71_51/g_samples_514_errs.npz"
```

```python
ds = load_dataset(fpath)
```

```python
g1p = ds['plus']['g'][:, :, 0]
g1p.shape
```

```python
g1m = ds['minus']['g'][:, :, 0]
g1m.shape
```

```python
m = (g1p.mean(1) - g1m.mean(1)) / 2/ 0.02  - 1
m.shape
```

```python
plt.hist(m)
```

```python
m.mean()  / 1e-3 
```

```python
m.std() / np.sqrt(len(m)) / 1e-3
```

```python
#  c
g2p = ds['plus']['g'][:, :, 1]
g2m = ds['minus']['g'][:, :, 1]
g2p.shape, g2m.shape
```

```python
cs = (g2p.mean(1) + g2m.mean(1)) / 2 
cs.mean() /1e-3, cs.std() / np.sqrt(500) /1e-3
```
