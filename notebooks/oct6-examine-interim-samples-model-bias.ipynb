```python
import matplotlib.pyplot as plt 
from bpd.io import load_dataset
import numpy as np
```

# Simple

# Multiple SLURM output

```python
samples = {} 
for ii in range(4):
    path =f"/pscratch/sd/i/imendoza/data/cache_chains/exp91_51/interim_samples_9920{ii}_plus.npz"
    ds = load_dataset(path)
    for p in ds['samples']:
        if p in samples:
            samples[p].append(ds['samples'][p])
        else: 
            samples[p] = [ds['samples'][p]]
```

```python
for p in samples:
    samples[p] = np.concatenate(samples[p], axis=0)
```

```python
samples['e1'].shape, samples.keys()
```

```python
ref_path = "/pscratch/sd/i/imendoza/data/cache_chains/exp72_51/interim_samples_51100_plus.npz"
ds_ref = load_dataset(ref_path)
ds_ref['samples']['e1'].shape
```

```python

e1 = samples['e1']
e2 = samples['e2']
lf = samples['lf']
e1.shape
```

```python
_, bins, _ = plt.hist(e1.flatten(), bins=31, histtype='step', density=True);
plt.hist(ds_ref['samples']['e1'].flatten(), bins=bins, histtype='step', density=True);
# plt.hist(e2.flatten(), bins=bins, histtype='step')
```

```python
_, bins, _ = plt.hist(e2.flatten(), bins=31, histtype='step', density=True);
plt.hist(ds_ref['samples']['e2'].flatten(), bins=bins, histtype='step', density=True);
```

```python
_, bins, _ = plt.hist(lf.flatten(), bins=25, histtype='step', density=True)
plt.hist(ds_ref['samples']['lf'].flatten(), bins=bins, histtype='step', density=True)
```
